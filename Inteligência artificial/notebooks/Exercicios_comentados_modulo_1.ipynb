#Código do 1° exercício comentado: Quicksort



```python
import numpy as np
```

```python
def quicksort(arr):
  if len(arr) <=1:
    return arr
  else:
    median = np.median([arr[0], arr[len(arr)//2], arr[-1]])
    pivot_index = np.where(arr == median)[0][0]

    arr[pivot_index], arr[-1] = arr[-1], arr[pivot_index]

    i=0
    for j in range(len(arr) - 1):
      if arr[j] <= arr[-1]:
        arr[i], arr[j] = arr[j], arr[i]
        i+=1

    arr[i], arr[-1] = arr[-1], arr[i]

    left = quicksort(arr[:i])
    right = quicksort(arr[i+1:])

    return np.concatenate((left, np.array([arr[i]]), right))
```

```python
arr = np.array([17, 21, 7, 3, 1, 6, 8])
arr_ordenado = quicksort(arr)
print(arr_ordenado)
```

#Código do Exercício Comentado 2: Pandas e label encoding

No pandas, há várias maneiras de se utilizar a técnica de label encoding, vou mostrar aqui duas das mais usadas.

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Sleep_health_and_lifestyle_dataset.csv")
```

## Utilizando pd.Categorical

Aqui, temos um exemplo utilizando o método categorical do pandas, onde ele recebe o atributo que desejamos codificar e cria um código numérico automaticamente.

```python
df['Gender_codes'] = pd.Categorical(df['Gender']).codes
```

```python
df[['Gender', 'Gender_codes']]
```

##Utilizando a função de mapping customizável

Aqui temos outra forma de codificar as categorias, sendo um trabalho mais manual e dando mais liberdade ao programador. Com ela, conseguimos dizer exatamente quais valores serão associados a cada categoria. O mais usual seria utilizar 0 e 1 para variáveis dicotômicas, utilizei outros números para mostrar que temos essa liberdade, desde que os valores sejam distintos.

```python
mapping = {'Male': 50, 'Female': 100}
df['Gender_codes'] = df['Gender'].map(mapping)
```

```python
df[['Gender', 'Gender_codes']]
```
